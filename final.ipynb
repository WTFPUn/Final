```python
def popset(url: str):
  pop = pd.read_csv(url, names=['8','7','province','0', '1', '6', '2', '3', '4', 'femalePop', 'malePop', 'totalPop', 'totalHouse','5'], thousands=',')
  pop = pop.iloc[1:,:]
  pop['province'] = pop['province'].str.replace('จังหวัด', '').str.replace(" ", '')
  pop = pop[['province', 'femalePop', 'malePop', 'totalPop', 'totalHouse']]
  return pop.merge(ProTHEN, how='left', left_on='province', right_on='provinceth')[['province_y', 'femalePop', 'malePop', 'totalPop', 'totalHouse']]
```
